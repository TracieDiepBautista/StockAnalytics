```python
import pandas as pd
from sqlalchemy import create_engine
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from datetime import datetime
```

```python
# Extract the data set from csv file: 
stock = "resources/all_stocks_5yr.csv"
allStock_df = pd.read_csv(stock)
allStock_df.head(10)
```

```python
# extract all data for most recent 5 years ()
```

```python
New_allStock_df = allStock_df.sort_values("date", ascending = False)
New_allStock_df
```

```python
groupstock = New_allStock_df.groupby("Name").mean()
groupstock
```

```python
top10stock = groupstock.sort_values("close", ascending=False)
top10stock.head(10)
```

```python
topstockvolume = groupstock.nlargest(10,"volume")
topstockvolume.head(10)
```

```python
# connect to local database: 
connection = "postgres:Postgres@localhost:5432/stockPandas_db"
engine = create_engine(f'postgresql://{connection}')
```

```python
# put df into sql database format

allStock_df.to_sql(name='stock_master', con=engine, if_exists='append', index=False)
```
